```python
import pandas as pd
```

# Accesando a archivo alojado en S3

```python
import boto3 as boto3

# Crear una sesión usando el perfil deseado
session = boto3.Session(profile_name='querier')

# Crear un cliente de S3 usando esa sesión
s3 = session.client('s3')
```

```python
# Nombre del bucket y del archivo
bucket_name = 'mydatatestobjects'
file_key = 'bestsellers-with-categories.csv'

# Leer el archivo CSV desde S3
obj = s3.get_object(Bucket=bucket_name, Key=file_key)
data = pd.read_csv(obj['Body'])

# Mostrar el DataFrame
data.head(2)
```
